### 🔴 Intervention 1: Repayment Burden Relief Programme

**Finding:** Customers with a repayment burden above 60% of income churn at nearly double the average rate.

**Recommended Action:**
- Proactively identify customers where `repayment_burden > 0.6` and offer loan restructuring or repayment holiday options.
- Automate flag in the CRM to trigger a call from a relationship manager when the threshold is crossed.

**Target segment:** ~420 customers currently in the High/Critical repayment burden bands.

---

### 🟡 Intervention 2: Income-Segmented Retention Offers

**Finding:** Churners earn ~52% less per month than active customers. Low-income customers have fewer products and thinner balances, making them price-sensitive.

**Recommended Action:**
- Introduce a tiered "Hifadhi" savings incentive for customers earning below KSH 50,000/month — e.g. bonus interest rate, free M-Pesa integration, or reduced fees.
- Align with the `monthly_income_ksh < 50,000` flag to personalise outreach.

---

### 🟠 Intervention 3: Regional Retention Campaigns in Nyeri & Eldoret

**Finding:** Nyeri (15.7%) and Eldoret (15.3%) both significantly exceed the 13.4% average.

**Recommended Action:**
- Deploy targeted branch-level retention campaigns in these two regions.
- Investigate whether local competitors (Equity Bank, KCB, M-Pesa agents) have launched recent promotions.
- Consider assigning dedicated relationship managers to high-value customers in these regions.

### 📊 Risk Scoring Framework

A simple 3-variable scorecard can be implemented in SQL or Excel with no ML infrastructure:

| Variable | Band | Score |
|----------|------|-------|
| `repayment_burden` | > 60% | +3 |
| `monthly_income_ksh` | < KSH 50,000 | +2 |
| `region` | Nyeri or Eldoret | +1 |
| `has_loan` | Yes | +1 |
| `credit_score` | < 600 | +1 |

**Risk tier:** Score 0–2 = Low, 3–4 = Medium, 5–6 = High, 7+ = Critical.

The retention team reviews all High and Critical customers monthly and prioritises outreach accordingly.
